```python
# ============================================================
# ¿Qué es Matplotlib?
# ============================================================
# Matplotlib es una biblioteca de Python utilizada para crear gráficos y visualizaciones.
# Es muy flexible y permite generar diferentes tipos de gráficos como líneas, dispersión,
# barras, tortas, histogramas, entre otros. También es ideal para personalizar cada
# aspecto del gráfico, como colores, estilos y etiquetas.

# ============================================================
# Instalación
# ============================================================
# Para usar Matplotlib, asegúrate de instalarla usando:
# !pip install matplotlib
# Si estás en Google Colab, simplemente importa la biblioteca directamente.

# ============================================================
# Importamos las bibliotecas necesarias
# ============================================================
import matplotlib.pyplot as plt  # Biblioteca principal para gráficos
import numpy as np  # Biblioteca auxiliar para manejar datos numéricos

# ============================================================
# 1. Gráfico de líneas básico
# ============================================================
# Los gráficos de líneas se utilizan para mostrar tendencias o cambios en los datos.

# Creamos los datos: listas de valores para los ejes X e Y
x = [0, 1, 2, 3, 4]  # Eje X: posiciones en el gráfico
y = [0, 1, 4, 9, 16]  # Eje Y: valores correspondientes (cuadrados)

# Creamos el gráfico de líneas
plt.plot(x, y, label="Cuadrados", color="blue", linestyle="--", marker="o")
# `label`: texto que aparecerá en la leyenda
# `color`: color de la línea
# `linestyle`: tipo de línea (en este caso, discontinua)
# `marker`: marcador en cada punto de datos

# Agregamos detalles al gráfico
plt.title("Gráfico de líneas básico")  # Título del gráfico
plt.xlabel("Eje X")  # Etiqueta para el eje X
plt.ylabel("Eje Y (Cuadrados)")  # Etiqueta para el eje Y
plt.legend()  # Muestra la leyenda en el gráfico
plt.grid(True)  # Activa la cuadrícula para facilitar la lectura

# Mostramos el gráfico
plt.show()

# ============================================================
# 2. Gráfico de dispersión
# ============================================================
# Los gráficos de dispersión muestran relaciones entre dos conjuntos de datos.

# Generamos datos aleatorios usando numpy
x_scatter = np.random.rand(50)  # 50 valores aleatorios entre 0 y 1 para X
y_scatter = np.random.rand(50)  # 50 valores aleatorios entre 0 y 1 para Y
colors = np.random.rand(50)  # Colores aleatorios
sizes = np.random.rand(50) * 1000  # Tamaños de puntos escalados

# Creamos el gráfico de dispersión
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.5, cmap='viridis')
# `c`: define los colores de los puntos
# `s`: define los tamaños de los puntos
# `alpha`: transparencia de los puntos (0 = transparente, 1 = opaco)
# `cmap`: mapa de colores usado para el gráfico

# Agregamos detalles al gráfico
plt.colorbar()  # Barra de colores para mostrar la escala
plt.title("Gráfico de dispersión")  # Título del gráfico
plt.xlabel("Eje X")  # Etiqueta para el eje X
plt.ylabel("Eje Y")  # Etiqueta para el eje Y

# Mostramos el gráfico
plt.show()

# ============================================================
# 3. Gráfico de barras
# ============================================================
# Los gráficos de barras se usan para comparar valores entre diferentes categorías.

# Datos categóricos
categorias = ['A', 'B', 'C', 'D']  # Categorías en el eje X
valores = [3, 7, 5, 9]  # Valores asociados a las categorías

# Creamos el gráfico de barras
plt.bar(categorias, valores, color='skyblue')  # Las barras serán de color azul claro

# Agregamos detalles al gráfico
plt.title("Gráfico de barras")  # Título del gráfico
plt.xlabel("Categorías")  # Etiqueta para el eje X
plt.ylabel("Valores")  # Etiqueta para el eje Y

# Mostramos el gráfico
plt.show()

# ============================================================
# 4. Gráfico de torta
# ============================================================
# Los gráficos de torta son ideales para mostrar proporciones.

# Datos para el gráfico
tamaños = [25, 35, 20, 20]  # Porcentajes de cada segmento
etiquetas = ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos']  # Etiquetas de las porciones
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Colores de las porciones
explode = (0.1, 0, 0, 0)  # Separamos la primera porción del resto

# Creamos el gráfico de torta
plt.pie(tamaños, explode=explode, labels=etiquetas, colors=colores,
        autopct='%1.1f%%', shadow=True, startangle=140)
# `explode`: desplaza las porciones hacia afuera
# `labels`: etiquetas para cada porción
# `autopct`: muestra el porcentaje dentro de las porciones
# `shadow`: agrega sombra al gráfico
# `startangle`: ángulo inicial del gráfico

plt.title("Gráfico de torta")  # Título del gráfico
plt.axis('equal')  # Asegura que el gráfico sea un círculo perfecto

# Mostramos el gráfico
plt.show()

# ============================================================
# 5. Subgráficos
# ============================================================
# Los subgráficos permiten mostrar múltiples gráficos en una misma figura.

# Generamos datos para seno y coseno
x = np.linspace(0, 10, 100)  # Genera 100 valores entre 0 y 10
y1 = np.sin(x)  # Seno de X
y2 = np.cos(x)  # Coseno de X

# Creamos un área para dos subgráficos
plt.subplot(2, 1, 1)  # Primer subgráfico: 2 filas, 1 columna, posición 1
plt.plot(x, y1, label="Seno", color="blue")
plt.title("Seno y Coseno (Subgráficos)")  # Título del primer subgráfico
plt.legend()  # Muestra la leyenda

plt.subplot(2, 1, 2)  # Segundo subgráfico: 2 filas, 1 columna, posición 2
plt.plot(x, y2, label="Coseno", color="green")
plt.legend()  # Muestra la leyenda en el segundo gráfico

# Ajustamos automáticamente los espacios entre gráficos
plt.tight_layout()

# Mostramos los subgráficos
plt.show()
```

Los graficos en Mathplotlib se dividen en 6 tipos y son los siguientes:

- grafico de lineas
- grafico de barras
- grafico de dispersión
- grafico de histograma
- grafico de pastel
- grafico de areas

A continuación, se muestra un ejemplo para cada tipo de grafico:

```python
#GRAFICO DE LINEAS
import matplotlib.pyplot as plt

# Datos: Ventas mensuales en una tienda
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
ventas = [3000, 3200, 2900, 3400, 3600, 3700]

# Gráfico de líneas
plt.plot(meses, ventas, color="blue", marker="o", linestyle="-", label="Ventas")
plt.title("Ventas mensuales")  # Título
plt.xlabel("Meses")  # Etiqueta del eje X
plt.ylabel("Ingresos ($)")  # Etiqueta del eje Y
plt.legend()  # Leyenda
plt.grid(True)  # Cuadrícula
plt.show()
```

```python
#GRÁFICO DE BARRAS
import matplotlib.pyplot as plt

# Datos: Número de estudiantes por carrera
carreras = ["Ingeniería", "Medicina", "Derecho", "Arquitectura", "Diseño"]
estudiantes = [120, 95, 85, 100, 75]

# Gráfico de barras
plt.bar(carreras, estudiantes, color="green")
plt.title("Estudiantes por carrera")  # Título
plt.xlabel("Carreras")  # Etiqueta del eje X
plt.ylabel("Número de estudiantes")  # Etiqueta del eje Y
plt.show()
```

```python
#GRÁFICO DE DISPERSIÓN
import matplotlib.pyplot as plt
import numpy as np

# Datos: Relación entre horas de estudio y puntajes en un examen
horas_estudio = [1, 2, 3, 4, 5, 6, 7, 8]
puntajes = [50, 55, 60, 65, 70, 75, 80, 85]

# Gráfico de dispersión
plt.scatter(horas_estudio, puntajes, color="red", label="Estudiantes")
plt.title("Relación entre horas de estudio y puntajes")  # Título
plt.xlabel("Horas de estudio")  # Etiqueta del eje X
plt.ylabel("Puntaje en el examen")  # Etiqueta del eje Y
plt.legend()  # Leyenda
plt.grid(True)  # Cuadrícula
plt.show()
```

```python
#GRÁFICO DE HISTOGRAMA
import matplotlib.pyplot as plt
import numpy as np

# Datos: Edades de 50 personas
edades = [18, 19, 21, 22, 22, 23, 25, 27, 29, 30, 32, 35, 36, 40, 45, 50, 51, 53, 55, 60]

# Gráfico de histograma
plt.hist(edades, bins=5, color="purple", edgecolor="black")  # `bins` define el número de categorías
plt.title("Distribución de edades")  # Título
plt.xlabel("Edades")  # Etiqueta del eje X
plt.ylabel("Número de personas")  # Etiqueta del eje Y
plt.show()
```

```python
#GRÁFICO DE PASTEL
import matplotlib.pyplot as plt

# Datos: Porcentaje de uso de redes sociales
redes_sociales = ["Facebook", "Instagram", "Twitter", "TikTok"]
uso = [40, 30, 20, 10]

# Gráfico de pastel
plt.pie(uso, labels=redes_sociales, autopct="%1.1f%%", startangle=90, colors=["blue", "pink", "yellow", "green"])
plt.title("Uso de redes sociales en porcentaje")  # Título
plt.axis("equal")  # Asegura que sea un círculo perfecto
plt.show()
```

```python
#GRÁFICO DE ÁREAS
import matplotlib.pyplot as plt

# Datos: Producción mensual de tres fábricas
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
fabrica_a = [100, 120, 110, 140, 150, 160]
fabrica_b = [90, 100, 95, 130, 135, 145]
fabrica_c = [80, 85, 88, 120, 125, 130]

# Gráfico de áreas apiladas
plt.stackplot(meses, fabrica_a, fabrica_b, fabrica_c, labels=["Fábrica A", "Fábrica B", "Fábrica C"], colors=["skyblue", "lightgreen", "pink"])
plt.title("Producción mensual por fábrica")  # Título
plt.xlabel("Meses")  # Etiqueta del eje X
plt.ylabel("Producción (unidades)")  # Etiqueta del eje Y
plt.legend(loc="upper left")  # Leyenda
plt.show()
```
