```python
# ============================================================
# Qu es Matplotlib?
# ============================================================
# Matplotlib es una biblioteca de Python utilizada para crear grficos y visualizaciones.
# Es muy flexible y permite generar diferentes tipos de grficos como lneas, dispersin,
# barras, tortas, histogramas, entre otros. Tambin es ideal para personalizar cada
# aspecto del grfico, como colores, estilos y etiquetas.

# ============================================================
# Instalacin
# ============================================================
# Para usar Matplotlib, asegrate de instalarla usando:
# !pip install matplotlib
# Si ests en Google Colab, simplemente importa la biblioteca directamente.

# ============================================================
# Importamos las bibliotecas necesarias
# ============================================================
import matplotlib.pyplot as plt  # Biblioteca principal para grficos
import numpy as np  # Biblioteca auxiliar para manejar datos numricos

# ============================================================
# 1. Grfico de lneas bsico
# ============================================================
# Los grficos de lneas se utilizan para mostrar tendencias o cambios en los datos.

# Creamos los datos: listas de valores para los ejes X e Y
x = [0, 1, 2, 3, 4]  # Eje X: posiciones en el grfico
y = [0, 1, 4, 9, 16]  # Eje Y: valores correspondientes (cuadrados)

# Creamos el grfico de lneas
plt.plot(x, y, label="Cuadrados", color="blue", linestyle="--", marker="o")
# `label`: texto que aparecer en la leyenda
# `color`: color de la lnea
# `linestyle`: tipo de lnea (en este caso, discontinua)
# `marker`: marcador en cada punto de datos

# Agregamos detalles al grfico
plt.title("Grfico de lneas bsico")  # Ttulo del grfico
plt.xlabel("Eje X")  # Etiqueta para el eje X
plt.ylabel("Eje Y (Cuadrados)")  # Etiqueta para el eje Y
plt.legend()  # Muestra la leyenda en el grfico
plt.grid(True)  # Activa la cuadrcula para facilitar la lectura

# Mostramos el grfico
plt.show()

# ============================================================
# 2. Grfico de dispersin
# ============================================================
# Los grficos de dispersin muestran relaciones entre dos conjuntos de datos.

# Generamos datos aleatorios usando numpy
x_scatter = np.random.rand(50)  # 50 valores aleatorios entre 0 y 1 para X
y_scatter = np.random.rand(50)  # 50 valores aleatorios entre 0 y 1 para Y
colors = np.random.rand(50)  # Colores aleatorios
sizes = np.random.rand(50) * 1000  # Tamaos de puntos escalados

# Creamos el grfico de dispersin
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.5, cmap='viridis')
# `c`: define los colores de los puntos
# `s`: define los tamaos de los puntos
# `alpha`: transparencia de los puntos (0 = transparente, 1 = opaco)
# `cmap`: mapa de colores usado para el grfico

# Agregamos detalles al grfico
plt.colorbar()  # Barra de colores para mostrar la escala
plt.title("Grfico de dispersin")  # Ttulo del grfico
plt.xlabel("Eje X")  # Etiqueta para el eje X
plt.ylabel("Eje Y")  # Etiqueta para el eje Y

# Mostramos el grfico
plt.show()

# ============================================================
# 3. Grfico de barras
# ============================================================
# Los grficos de barras se usan para comparar valores entre diferentes categoras.

# Datos categricos
categorias = ['A', 'B', 'C', 'D']  # Categoras en el eje X
valores = [3, 7, 5, 9]  # Valores asociados a las categoras

# Creamos el grfico de barras
plt.bar(categorias, valores, color='skyblue')  # Las barras sern de color azul claro

# Agregamos detalles al grfico
plt.title("Grfico de barras")  # Ttulo del grfico
plt.xlabel("Categoras")  # Etiqueta para el eje X
plt.ylabel("Valores")  # Etiqueta para el eje Y

# Mostramos el grfico
plt.show()

# ============================================================
# 4. Grfico de torta
# ============================================================
# Los grficos de torta son ideales para mostrar proporciones.

# Datos para el grfico
tamaos = [25, 35, 20, 20]  # Porcentajes de cada segmento
etiquetas = ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos']  # Etiquetas de las porciones
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Colores de las porciones
explode = (0.1, 0, 0, 0)  # Separamos la primera porcin del resto

# Creamos el grfico de torta
plt.pie(tamaos, explode=explode, labels=etiquetas, colors=colores,
        autopct='%1.1f%%', shadow=True, startangle=140)
# `explode`: desplaza las porciones hacia afuera
# `labels`: etiquetas para cada porcin
# `autopct`: muestra el porcentaje dentro de las porciones
# `shadow`: agrega sombra al grfico
# `startangle`: ngulo inicial del grfico

plt.title("Grfico de torta")  # Ttulo del grfico
plt.axis('equal')  # Asegura que el grfico sea un crculo perfecto

# Mostramos el grfico
plt.show()

# ============================================================
# 5. Subgrficos
# ============================================================
# Los subgrficos permiten mostrar mltiples grficos en una misma figura.

# Generamos datos para seno y coseno
x = np.linspace(0, 10, 100)  # Genera 100 valores entre 0 y 10
y1 = np.sin(x)  # Seno de X
y2 = np.cos(x)  # Coseno de X

# Creamos un rea para dos subgrficos
plt.subplot(2, 1, 1)  # Primer subgrfico: 2 filas, 1 columna, posicin 1
plt.plot(x, y1, label="Seno", color="blue")
plt.title("Seno y Coseno (Subgrficos)")  # Ttulo del primer subgrfico
plt.legend()  # Muestra la leyenda

plt.subplot(2, 1, 2)  # Segundo subgrfico: 2 filas, 1 columna, posicin 2
plt.plot(x, y2, label="Coseno", color="green")
plt.legend()  # Muestra la leyenda en el segundo grfico

# Ajustamos automticamente los espacios entre grficos
plt.tight_layout()

# Mostramos los subgrficos
plt.show()
```

Los graficos en Mathplotlib se dividen en 6 tipos y son los siguientes:

- grafico de lineas
- grafico de barras
- grafico de dispersin
- grafico de histograma
- grafico de pastel
- grafico de areas

A continuacin, se muestra un ejemplo para cada tipo de grafico:

```python
#GRAFICO DE LINEAS
import matplotlib.pyplot as plt

# Datos: Ventas mensuales en una tienda
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
ventas = [3000, 3200, 2900, 3400, 3600, 3700]

# Grfico de lneas
plt.plot(meses, ventas, color="blue", marker="o", linestyle="-", label="Ventas")
plt.title("Ventas mensuales")  # Ttulo
plt.xlabel("Meses")  # Etiqueta del eje X
plt.ylabel("Ingresos ($)")  # Etiqueta del eje Y
plt.legend()  # Leyenda
plt.grid(True)  # Cuadrcula
plt.show()
```

```python
#GRFICO DE BARRAS
import matplotlib.pyplot as plt

# Datos: Nmero de estudiantes por carrera
carreras = ["Ingeniera", "Medicina", "Derecho", "Arquitectura", "Diseo"]
estudiantes = [120, 95, 85, 100, 75]

# Grfico de barras
plt.bar(carreras, estudiantes, color="green")
plt.title("Estudiantes por carrera")  # Ttulo
plt.xlabel("Carreras")  # Etiqueta del eje X
plt.ylabel("Nmero de estudiantes")  # Etiqueta del eje Y
plt.show()
```

```python
#GRFICO DE DISPERSIN
import matplotlib.pyplot as plt
import numpy as np

# Datos: Relacin entre horas de estudio y puntajes en un examen
horas_estudio = [1, 2, 3, 4, 5, 6, 7, 8]
puntajes = [50, 55, 60, 65, 70, 75, 80, 85]

# Grfico de dispersin
plt.scatter(horas_estudio, puntajes, color="red", label="Estudiantes")
plt.title("Relacin entre horas de estudio y puntajes")  # Ttulo
plt.xlabel("Horas de estudio")  # Etiqueta del eje X
plt.ylabel("Puntaje en el examen")  # Etiqueta del eje Y
plt.legend()  # Leyenda
plt.grid(True)  # Cuadrcula
plt.show()
```

```python
#GRFICO DE HISTOGRAMA
import matplotlib.pyplot as plt
import numpy as np

# Datos: Edades de 50 personas
edades = [18, 19, 21, 22, 22, 23, 25, 27, 29, 30, 32, 35, 36, 40, 45, 50, 51, 53, 55, 60]

# Grfico de histograma
plt.hist(edades, bins=5, color="purple", edgecolor="black")  # `bins` define el nmero de categoras
plt.title("Distribucin de edades")  # Ttulo
plt.xlabel("Edades")  # Etiqueta del eje X
plt.ylabel("Nmero de personas")  # Etiqueta del eje Y
plt.show()
```

```python
#GRFICO DE PASTEL
import matplotlib.pyplot as plt

# Datos: Porcentaje de uso de redes sociales
redes_sociales = ["Facebook", "Instagram", "Twitter", "TikTok"]
uso = [40, 30, 20, 10]

# Grfico de pastel
plt.pie(uso, labels=redes_sociales, autopct="%1.1f%%", startangle=90, colors=["blue", "pink", "yellow", "green"])
plt.title("Uso de redes sociales en porcentaje")  # Ttulo
plt.axis("equal")  # Asegura que sea un crculo perfecto
plt.show()
```

```python
#GRFICO DE REAS
import matplotlib.pyplot as plt

# Datos: Produccin mensual de tres fbricas
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
fabrica_a = [100, 120, 110, 140, 150, 160]
fabrica_b = [90, 100, 95, 130, 135, 145]
fabrica_c = [80, 85, 88, 120, 125, 130]

# Grfico de reas apiladas
plt.stackplot(meses, fabrica_a, fabrica_b, fabrica_c, labels=["Fbrica A", "Fbrica B", "Fbrica C"], colors=["skyblue", "lightgreen", "pink"])
plt.title("Produccin mensual por fbrica")  # Ttulo
plt.xlabel("Meses")  # Etiqueta del eje X
plt.ylabel("Produccin (unidades)")  # Etiqueta del eje Y
plt.legend(loc="upper left")  # Leyenda
plt.show()
```
