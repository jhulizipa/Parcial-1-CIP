```python
from google.colab import drive
drive.mount('/content/drive/')
import pandas as pd
import plotly.express as px

path_file= "/content/drive/MyDrive/CIP/Datos.csv"
df=pd.read_csv(path_file, sep=",", header=0, encoding='latin-1')

print(df.shape)
df.columns

```

```python
df.columns = ['Id', 'Apellidos', 'Nombre', 'Género', 'Edad', 'Fecha de Ingreso', 'Puesto', 'Ubicación', 'Sueldo Base', 'Bonificación', 'Total']
df.columns
```

```python
df[ 'Género' ].value_counts()
```

```python
df.sort_values('Total')
```

```python
import matplotlib.pyplot as plt

x_values = df['Género'].unique()
y_values = df['Género'].value_counts(dropna=False).loc[x_values].tolist()

x_values = [str(val) for val in x_values]

plt.bar(x_values, y_values)
plt.show()
plt.close('all')
```
